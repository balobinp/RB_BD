```python
def per_cals(x):
    return round((x['SUBSCRIBER_ID_04'] - x['SUBSCRIBER_ID_03']) * 100 / x['SUBSCRIBER_ID_04'], 1)
```

```python
# by usage_type

group = ['USAGE_TYPE']

df_usage_temp = pd.merge(df_usage_03.groupby(group, as_index=False)['SUBSCRIBER_ID'].count(),
         df_usage_04.groupby(group, as_index=False)['SUBSCRIBER_ID'].count(),
         how='outer', on=group, suffixes=['_03', '_04'])
df_usage_temp['INCR'] = df_usage_temp.apply(per_cals, axis=1)
df_usage_temp.rename({'SUBSCRIBER_ID_03':'COUNT_USAGE_03', 'SUBSCRIBER_ID_04':'COUNT_USAGE_04'}, axis=1, inplace=True)
df_usage_temp
```

```python
from pandas import Series, DataFrame
import numpy as np
import pandas as pd
df=DataFrame({'data1':np.random.randint(0,10,size=10),'data2':np.random.randint(0,10,size=10),
              'key1':list('ab'*5),'key2':list('fg'*5)},index=(list('klmno'*2)))
display(df)
df.groupby(['key1'], as_index=False)['data1'].max() # Группировка по полю
df.groupby(level=0)['data1'].max().reset_index() # Группировка по индексу
def unique_val(a):
    return a.nunique()
def count_val(a):
    return a.count()
df.groupby(['key1'], as_index=False)['data1'].agg([sum, max, unique_val, count_val]) # Применение нескольких функций
```

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
a=np.array([[5,'one'], [2,'one'], [3,'one'], [7,'two'],
            [2,'two'], [1,'two'], [70,'three'], [30,'three']])
c=np.array(['one','desc'])
frame = DataFrame(a,columns=c)

frame['one'] = pd.to_numeric(frame['one'], errors='coerce',downcast='float')
display(frame)
def test(gr):
    r=gr.one.astype(float) #значению r присваиваются значения группы
    gr['prop']=r/r.sum() #добавляем поле rpot и присваиваем ему значение
    return gr
frame_gr=frame.groupby(['desc']).apply(test)
frame_gr
```
